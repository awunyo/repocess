```python
plt.scatter(data['fixed acidity'], data['citric acid'])
plt.title('Fixed Acidity vs Citric Acid ')
plt.xlabel('Fixed Acidity')
plt.ylabel('Citric Acid')
plt.savefig(os.getcwd() + '/graph1.png')
plt.show()
```

```python
plt.scatter(data['alcohol'], data['pH'])
plt.title('Alcohol vs pH')
plt.xlabel('Alcohol')
plt.ylabel('pH')
plt.savefig(os.getcwd() + '/graph2.png')
plt.show()
```

### Box Plot: Residual Sugar Distribution
This box plot offers a summary of the distribution of 'Residual Sugar' within the dataset. The y-axis displays the values of 'Residual Sugar'. The box encompasses the interquartile range , with a central line denoting the median. Whiskers extend to represent the overall data range, and any data points beyond the whiskers are identified as outliers. The title and y-axis label provide additional context to aid in the interpretation of the plot.
